```python
# import all necessary libraries and methods
import pandas as pan
import numpy as num
```

```python
# Task1: create a csv file using excel and saved as Project1.csv
# Task2: read CSV into Pandas
twoairlines = pan.read_csv ('Project1.csv')
```

```python
# Task2.1: Check to see the output is correct
twoairlines
```

```python
# Task3.1: Use sum(function) to get a total/summary of flights per arilines based on each city.
# Use "axis = 1" to appends a column to Dataframes.
twoairlines['SUM'] = twoairlines.sum (axis=1)
```

```python
# Task3.2: Check 
twoairlines
```

```python
# Task3.3: Add a column where the flight delayed is shown in percentage by using ".div"
twoairlines['Delayed_Percentage'] = ( twoairlines['Delayed'].div(twoairlines['SUM']) * 100 )
```

```python
# Task3.4: Check
twoairlines
```

```python
# Task3.5: For looks, will round it to One Decimal
twoairlines['Delayed_Percentage'] = twoairlines['Delayed_Percentage'].round(decimals=1)
```

```python
# Task3.6: Check
twoairlines
```

```python
# Task3.7: Arrange the order of Delayed_Percentage column
twoairlines.sort_values('Delayed_Percentage', ascending = False)
```
